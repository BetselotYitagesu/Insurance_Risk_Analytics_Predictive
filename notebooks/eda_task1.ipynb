```python
# # Drop Columns with Too Much Missing Data... given the previous outputs
df.drop(columns=['NumberOfVehiclesInFleet'], inplace=True)

# Fill missing with "Unknown" (or "Other" if preferred).

df['Bank'] = df['Bank'].fillna('Unknown')
df['AccountType'] = df['AccountType'].fillna('Unknown')

# Clean make, Model, Cylinders, kilowatts, cubiccapacity

df['make'] = df['make'].str.strip().fillna('Unknown')
df['Model'] = df['Model'].fillna('Unknown')
df['Cylinders'] = df['Cylinders'].fillna(-1)
df['kilowatts'] = df['kilowatts'].fillna(-1)
df['cubiccapacity'] = df['cubiccapacity'].fillna(-1)

# Fill NaN with 'Unknown' or 'No' (if you assume default is 'No').
df['CrossBorder'] = df['CrossBorder'].fillna('No')


df['VehicleIntroDate'] = df['VehicleIntroDate'].fillna('Unknown')  # if string/date
df['NumberOfDoors'] = df['NumberOfDoors'].fillna(-1)
df['mmcode'] = df['mmcode'].fillna('Unknown')
df['bodytype'] = df['bodytype'].fillna('Unknown')

```
